```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tkinter import Tk
from tkinter.filedialog import askopenfilename

# Fungsi untuk menampilkan gambar
def show_image(title, image):
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Fungsi untuk memilih file gambar dan memprosesnya
def select_and_process_image():
    # Membuka dialog untuk memilih file gambar
    Tk().withdraw()  # Menutup jendela Tkinter utama
    file_path = askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png")])
    
    if file_path:
        # Baca gambar
        image = cv2.imread(file_path)

        # Konversi gambar ke ruang warna HSV
        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        # Tentukan rentang warna yang ingin dideteksi (contoh: warna merah)
        lower_red = np.array([0, 120, 70])
        upper_red = np.array([10, 255, 255])

        # Buat masker untuk mendeteksi warna
        mask = cv2.inRange(hsv_image, lower_red, upper_red)

        # Terapkan masker ke gambar asli
        result = cv2.bitwise_and(image, image, mask=mask)

        # Tampilkan gambar asli, masker, dan hasil
        show_image('Original Image', image)
        show_image('Mask', mask)
        show_image('Detected Color', result)
    else:
        print("No file selected")

# Panggil fungsi untuk memilih dan memproses gambar
select_and_process_image()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tkinter import Tk
from tkinter.filedialog import askopenfilename
from tkinter import Button

# Fungsi untuk menampilkan gambar
def show_image(title, image):
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Fungsi untuk memilih file gambar dan memprosesnya
def select_and_process_image():
    # Membuka dialog untuk memilih file gambar
    Tk().withdraw()  # Menutup jendela Tkinter utama
    file_path = askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png")])
    
    if file_path:
        # Baca gambar
        image = cv2.imread(file_path)

        # Konversi gambar ke ruang warna HSV
        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        # Tentukan rentang warna yang ingin dideteksi (contoh: warna biru)
        lower_blue = np.array([100, 150, 0])
        upper_blue = np.array([140, 255, 255])

        # Buat masker untuk mendeteksi warna
        mask = cv2.inRange(hsv_image, lower_blue, upper_blue)

        # Terapkan masker ke gambar asli
        result = cv2.bitwise_and(image, image, mask=mask)

        # Tampilkan gambar asli, masker, dan hasil
        show_image('Original Image', image)
        show_image('Mask', mask)
        show_image('Detected Color', result)
    else:
        print("No file selected")

# Membuat tombol untuk memilih gambar
def create_upload_button():
    root = Tk()
    root.title("Upload Image for Color Detection")
    button = Button(root, text="Select Image", command=select_and_process_image)
    button.pack()
    root.mainloop()

# Panggil fungsi untuk membuat tombol unggah
create_upload_button()
```
